### #2 - Create a UDF <b>
<p>Using the Boston Red Sox data, create your own UDF which creates a new column called 'All-Star' and puts every player with either a batting average over .280 or an on base percentage of over .360 with a result of 'Yes' in the column and 'No' if not.</p>

```python
sox18 = pd.read_csv('../../files/redsox_2018_hitting.txt')

def star(row):
    ba = row['BA']
    obp = row['OBP']
    if ba >= .280 or obp >= .360:
        return 'Yes'
    else:
        return 'No'
    
sox18['AllStar'] = sox18.apply(star, axis = 1)

sox18[['Name', 'OBP', 'BA', 'AllStar']]
```

### #3 - Find the Mean for the Red Sox <br>
<p>Compare both years of Red Sox Hitting Data and diaplay the following</p>
<p>Get total number of home runs and rbi's</p>
<p>Based on the total number of HR's who had the better year? What about by total RBI's?</p>


```python
#Test to make sure you have the correct data
sox17 = pd.read_csv('../../files/redsox_2017_hitting.txt')
sox18 = pd.read_csv('../../files/redsox_2018_hitting.txt')

mean17 = sox17[['HR', 'RBI']].sum()
mean18 = sox18[['HR', 'RBI']].sum()

print(f'Total for 2017: {mean17}')
print(f'Total for 2018: {mean18}')

if mean17['HR'] > mean18['HR']:
    print('Better year based on HR is 2017')
else:
    print('Better year based on HR is 2018')
    
if mean17['RBI'] > mean18['RBI']:
    print('Better year based on RBI is 2017')
else:
    print('Better year based on RBI is 2018')
```
